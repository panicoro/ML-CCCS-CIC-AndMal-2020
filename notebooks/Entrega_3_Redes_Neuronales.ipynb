## **OBJETIVOS Y ALCANCES**

Como parte de nuestro trabajo integrador, nos interesamos por la temática de **Ciberseguridad** ya que es un tema de mucho interés en la actualidad. En particular queremos investigar sobre los programas maliciosos o *malware* y sus diferentes tipos.
Un programa malicioso o *malware* es cualquier tipo de software que realiza acciones dañinas en un sistema informático de forma intencionada (a diferencia del *software defectuoso*) y sin el conocimiento del usuario.
Ejemplos típicos de estas actividades maliciosas son el​ robo de información (p. ej. a través de troyanos), dañar o causar un malfuncionamiento del sistema informático, provocar perjuicios económicos, chantajear a propietarios de los datos de sistemas informáticos (p. ej. con un Ransomware), permitir el acceso de usuarios no autorizados, provocar molestias o una combinación de varias de estas actividades.

La importancia del estudio y la detección del malware de Android radica en la creciente amenaza que representa para la seguridad digital y la privacidad de los usuarios de dispositivos móviles. Con el aumento exponencial del uso de teléfonos inteligentes y dispositivos Android en todo el mundo, el riesgo de exposición a malware también ha aumentado significativamente. La detección efectiva del malware de Android es crucial para proteger la integridad de los datos personales, financieros y confidenciales almacenados en estos dispositivos. El malware puede causar daños graves, como robo de información sensible, acceso no autorizado a cuentas bancarias y tarjetas de crédito, espionaje digital y bloqueo del dispositivo, entre otros.

La detección de malware es esencialmente un problema de clasificación de software basado en la información recopilada del análisis del programa, por lo cual nos interesa desarrollar un **modelo de clasicación** que más allá de indetificar una categoria de malware nos ayude a detectar la potencial presencia de estas amenazas.

En este trabajo se encontró que algunas muestras de malware son ejecutadas sólo después de reiniciar el emulador. Por lo tanto, después de analizar las muestras y extraer las características, el emulador es rebooteado y el proceso se repite para extraer las características dinámicas de nuevo. Por lo tanto se capturaron 2 archivos CSV por cada categoría de malware:

* Antes de rebootear el emulador (before)
* Después de rebootear el emulador (after)

Un vista detallada de la arquitectura y los componentes usados para generar el dataset puede verse en la esta figura:

![image.png](https://i.postimg.cc/yxhn8PVZ/arquitecture.png)

Se configuraron 30 máquinas virtuales para analizar las muestras en forma paralela, de tal forma que hay 3 componentes principales:

1. **Base de datos**: como primer paso se entrega un malware desde el host por un comando SSH. Se almacena en la base de datos MongoDB y es manejado por un script, el cual enviará las muestras al siguiente componente para configurar su ejecución en el emulador. La base de datos es la responsable de almacenar los hashes y el status del análisis.
2. **Servidor de análisis (Sandbox Android)**: es el componente más importante que maneja las instancias de emulador, ejecuta análisis de individuales e invoca otros componentes. Una vez que un nuevo job es entregado, crea un directorio temporal y comienza a calcular los valores de las características dinámicas. Después de esto el emulador es reiniciado por el sandbox y captura todos los valores nuevamente. Finalmente almacena estos en los archivos before y after correspondientes, como se mencionó anteriormente.
4. **Emulador Android**: es responsable de emular el sistema operativo Android (OS), relacionarse con el servidor de análisis  y recolectar el tráfico de red. El ciclo de análisis de las muestras es controlado por el sandbox.

Además de los componentes descritos, existe una carpeta compartida para almacenar las muestras de malware y los resultados del análisis dinámico. Se menciona además que algunas muestras de malware no pudieron ejecutarse, ya sea porque no tenían entry point o porque terminaban abruptamente. Por lo que hubo menos datos en los .csv que los esperado al terminar el análisis.

### Variable: **Family**

Veamos ahora la cantidad de familias por cada categoria de malware.

````python
def get_family_by_category(data):
  return data.groupby(['Category',
                       'Family']).Memory_HeapSize.count()

def get_family_distrubution(data):
  family_by_category = get_family_by_category(malware_df)
  categories = family_by_category.index.get_level_values(level=0).unique()
  family_amount = [family_by_category[cat].count() for cat in categories]
  category_total = [family_by_category[cat].sum() for cat in categories]
  return pd.DataFrame({"Family amount": family_amount, "Total category": category_total},
                       index=categories)
family_distribution = get_family_distrubution(malware_df)
family_distribution
````

Comparando con la documentación dada en la web del dataset se observa que para algunas categorias aparecen mas familias que las que se esperan. Por ejemplo PUA fueron contabilizadas 9 familias pero en la documentacion se reportaron 8, como se ve mas abajo. A su vez en algunos casos, se tienen menos familias que las reportadas, por ejemplo Adware donde se reportaron 48 categorias pero hay 43. Esto último puede explicarse por el hecho de algunos malware no pudieron ejecutarse o se detuvieron al hacerlo como se explicó anteriormente.

![image.png](https://i.postimg.cc/YqF6z8YK/families.png)

Examinando las clases con más categorías que las reportadas, vemos que esto se explica por la existencia de un valor ```<unknown>```. Es decir varias categorias de malware tienen muestras que no tienen etiqueta familia. Veamos cuáles son estas categorías y contabilizemos las cantidad de  ```<unknown>``` sobre el total de cada categoria. Estos de alguna manera constituyen datos faltantes aunque no aparezcan como ```NaN```.

````python
family_by_category = get_family_by_category(malware_df)
categories = family_by_category.index.get_level_values(level=0).unique()
for v in family_by_category.index.values:
    if "<unknown>" in v[1]:
        print(v)
````

````python
def get_unknown_amount(group_by, cat):
    try:
        return group_by[cat]["<unknown>"]
    except KeyError:
        return 0

family_amount = family_distribution["Family amount"]
category_total = family_distribution["Total category"]
unknown_amount = [get_unknown_amount(family_by_category, cat) for cat in categories]
family_df = pd.DataFrame({"Family_amount": family_amount, "Total_category": category_total,
                          "Unknown_amount": unknown_amount}, index=categories)
family_df["% Unknown"] = np.round((family_df["Unknown_amount"] / family_df["Total_category"]) * 100, 2)
family_df
````

Con esto confirmamos que la unica categoria sin datos ```<unkwown>``` en las familias es Adware. El resto tiene algun porcentaje de datos desconocidos, pero es relativamente bajo y menor al 7% en la gran mayoria de los casos. Esto debe tenerse en cuenta en el futuro para saber como se va a manejar este valor si se quiere trabajar con la variable Family. Ademas esto podria estar hablandonos de
un nuevo tipo de malware dentro de cada cateogria.



Sobre la variable categoria *Family* habiamos visto que casi todas las categorias tenían un porcentaje de valores ```<unkown>```. Deberiamos ver tambien si los nombres de familia son mutuamente excluyentes entre categorias.  Es decir, comprobamos que el único valor compartido entre categorías es ```<unkwown>```


````python
family_by_category = malware_df.groupby(['Category',
                                         'Family']).Memory_HeapSize.count()
categories = malware_df["Category"].unique()
for cat in categories:
  set_family = set(family_by_category[cat].index)
  for cat_to_compare in categories:
    if cat != cat_to_compare:
        res = set_family & set(family_by_category[cat_to_compare].index)
        assert res != {"unknown"}
````

Notemos que tanto para *No_Category* como para *Zero_Day* se tienen una cantidad muy grande de familias: 335 y 2576, respectivamente.

#### Manejo de valores ```<unknown>``` y familias numerosas.


En base a los análisis realizados, se tomaron las siguientes decisiones:
1. Se excluirán las muestras con el label *No_Category*, porque no es una categoria de malware en si y tiene demasiadas clases de familia de malware.
2. Se excluirán las muestras de la categoría *Zero_Day* ya que tampoco tiene valores significativos en la categoria Family.
3. Se excluirán las muestras que tengan ```<unknown>``` como familia sin importar de que categoría sean.

````python
# Excluimos las 2 categorias mencionadas y los valores unknown
malware_df = malware_df.drop(malware_df[
    malware_df['Category'] == 'No_Category'].index)
malware_df = malware_df.drop(malware_df[
    malware_df['Category'] == 'Zero_Day'].index)
malware_df = malware_df.drop(
      malware_df[malware_df['Family'] == '<unknown>'].index)
````

## Modelo de Redes Neuronales

Realizaremos 2 modelos usando Multilayer Perceptron (MLP) uno para predecir la categoría de malware y otro para la familia. Para armar los modelos usaremos la libreria [Tensorflow](https://www.tensorflow.org/) y [Keras](https://keras.io/), por lo que procederemos a importarlas.

````python
import tensorflow as tf
import keras
print(tf.__version__)
import warnings
warnings.filterwarnings("ignore")
````

Para los modelos neuronales necesitamos transformar las variables objetivos que son categóricas, en numericas para que sean aptas para la red. Usamos *LabelEncoder* para ello.

````python
from sklearn.preprocessing import LabelEncoder

le_cat = LabelEncoder()
malware_df["Category_encod"] = le_cat.fit_transform(malware_df["Category"])
le_fam = LabelEncoder()
malware_df["Family_encod"] = le_fam.fit_transform(malware_df["Family"])

# Encoded values
y_cat = malware_df["Category_encod"].values
y_fam = malware_df["Family_encod"].values
````

### Modelo para predecir Category

Para las redes neuronales es importante escalar los datos del dataset, por lo que usaremos *StandardScaler* sobre el dataset y luego procederemos a separar el dataset con el ratio 70:30 para entrenar el modelo con el 70% y dejar el 30% como validación. Para la variable target, ```y_cat``` necesitamos convertir el vector en una matriz binaria, lo cual haremos usando la funcion *to_categorical* de keras.



````python
from sklearn.preprocessing import StandardScaler

X = df_feat_30

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
````

````python
from sklearn.model_selection import train_test_split

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y_cat,
                                                    test_size=0.3,
                                                    random_state=42,
                                                    stratify=y_cat)
````

````python
from keras.utils import to_categorical

y_train_bin = to_categorical(y_train)
y_test_bin = to_categorical(y_test)
````

Configuramos la red neuronal que usaremos, la cual tendrá 6 capas en total. La primera de 512 neuronas, 4 ocultas de de 256, 128, 64 y 32 y la salida de 12. Para todas las capas se uso la funcion de activación *ReLU* y en la de salida se se usó *Sofmax*. Además se agrego Dropout para desde la segunda capa en adelante como tecnica de regularización.

````python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.losses import categorical_crossentropy
from keras.optimizers import Adam

# Create the model
model = Sequential()
model.add(Dense(512, activation='relu', input_dim=30))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(12, activation='softmax'))
model.summary()

# Compile the model
model.compile(loss=categorical_crossentropy,
              optimizer=Adam(learning_rate=0.001),
              metrics=['accuracy', 'F1Score', 'Recall', 'Precision'])
````

Realizamos el ajusto del modelo con ```n_epochs = 30```  y un ```batch_size = 32```. En cada iteracion se evaula el modelo en un conjunto de datos de test que es el 30 % del total de los datos usadaos para entrenar.

````python
n_epochs = 30
random_state = 1234
batch_size = 32
verbosity = 1
validation_split = 0.30

# Fit data to model
history = model.fit(X_train, y_train_bin,
          batch_size=batch_size,
          epochs=n_epochs,
          verbose=verbosity,
          validation_split=validation_split)
````

````python
def plot_history(history):
  fig = plt.figure(figsize=(16,4))
  data_plot = [{"name": "accuracy",
                            "name_test": "val_accuracy",
                            "title": "Model accuracy",
                            "label": "Accuracy"},
               {"name": "loss",
                        "name_test": "val_loss",
                        "title": "Model loss",
                        "label": "Loss"}]
  for idx, dt in enumerate(data_plot):
    ax = fig.add_subplot(1, 2, idx+1)
    ax.plot(history.history[dt["name"]])
    ax.plot(history.history[dt["name_test"]])
    ax.set_title(dt["title"])
    ax.set_ylabel(dt["label"])
    ax.set_xlabel('Epoch')
    ax.legend(['Train', 'Test'], loc='upper left')
  fig.subplots_adjust(hspace=0.4, wspace=0.3)
  fig.suptitle('Accuracy and loss', fontsize=20)

plot_history(history)
# Generate generalization metrics
score = model.evaluate(X_test, y_test_bin, verbose=0)
print(f"Loss: {score[0]}")
for i in range(1, len(model.metrics_names)):
    print(f"* {model.metrics_names[i]}: {score[i]}")
````

### SMOTE

SMOTE (Synthetic Minority Over-sampling Technique) es una técnica utilizada para abordar el problema del desequilibrio de clases en conjuntos de datos, particularmente en contextos de clasificación. El desequilibrio de clases ocurre cuando una clase está significativamente subrepresentada en comparación con otras, lo que puede llevar a que los modelos de aprendizaje automático tengan un rendimiento sesgado hacia la clase mayoritaria.

La técnica realiza los siguientes pasos:
1. Identificación de Vecinos Más Cercanos. Para cada muestra en la clase minoritaria, se identifican sus *k* vecinos más cercanos utilizando una medida de distancia (generalmente la distancia euclidiana).
2. Generación de Muestras Sintéticas.Se selecciona aleatoriamente uno de los *k* vecinos más cercanos.Se genera una nueva muestra sintética interpolando entre la muestra original y su vecino seleccionado. Esto se hace tomando un punto aleatorio en la línea que conecta la muestra original y su vecino.
 Matemáticamente, si $x_i$ es la muestra original y $x_{zi}$  es uno de sus vecinos más cercanos, la nueva muestra sintética $x_{new}$ se calcula como:

$\begin{equation} x_{new} = x_i + rand(0,1) \times (x_{zi} - x_i) \end{equation}$

donde donde $rand(0,1)$ es un número aleatorio entre 0 y 1.

3. Repetición del Proceso: Este proceso se repite para muchas muestras de la clase minoritaria hasta que la clase minoritaria tenga el mismo número de muestras que la clase mayoritaria o el nivel de balance deseado.

**Ventajas de SMOTE**
* Mejora del Balance de Clases: Ayuda a equilibrar las clases, lo que puede mejorar el rendimiento de los modelos de clasificación.
* Generación de Datos Sintéticos: Al generar nuevas muestras sintéticas en lugar de simplemente replicar las existentes, SMOTE puede ayudar a prevenir el sobreajuste.

**Desventajas de SMOTE**
* Introducción de Ruido: Si los datos originales tienen ruido, SMOTE también puede generar muestras sintéticas que no representan bien la distribución subyacente.
* No Considera la Distribución Global: Al generar nuevas muestras sintéticas, SMOTE solo considera los vecinos más cercanos y no la distribución global de la clase minoritaria, lo que puede llevar a una representación inexacta de la distribución original.

Luego usaremos la funcion ```SMOTE``` de la libreria ```imblearn``` para aumentar la cantidad de datos en la muestra de entrenamiento. Veamos que luego se tienen la misma cantidad de muestras por clases. Se escalan los datos y se divide la muestra para entrenamiento y test.

````python
from imblearn.over_sampling import SMOTE
from collections import Counter

sm = SMOTE(random_state=42)

X_res, y_res = sm.fit_resample(X, y_cat)
resampled_distribution = Counter(y_res)
sorted_resampled_distribution = dict(sorted(resampled_distribution.items()))
print('Distribución de clases después del sobremuestreo:',
       sorted_resampled_distribution)

# Escalamos los datos
scaler = StandardScaler()
scaler.fit(X_res)
scaled_data_res = scaler.transform(X_res)
scaled_data_res = pd.DataFrame(scaled_data_res, columns=X_res.columns)

# Cambiamos
y_cat_bin = to_categorical(y_res)

# Dividir en conjunto de entrenamiento y prueba
X_res_train, X_res_test, y_res_train, y_res_test = train_test_split(
                                                    scaled_data_res, y_cat_bin,
                                                    test_size=0.3,
                                                    random_state=42,
                                                    stratify=y_cat_bin)
````

````python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.losses import categorical_crossentropy
from keras.optimizers import Adam
import random

seed_value = 1234
tf.random.set_seed(seed_value)
np.random.seed(seed_value)
random.seed(seed_value)

# Create the model
model_res = Sequential()
model_res.add(Dense(512, activation='relu', input_dim=30))
model_res.add(Dense(256, activation='relu'))
model_res.add(Dropout(0.3))
model_res.add(Dense(128, activation='relu'))
model_res.add(Dropout(0.2))
model_res.add(Dense(64, activation='relu'))
model_res.add(Dropout(0.1))
model_res.add(Dense(32, activation='relu'))
model_res.add(Dropout(0.05))
model_res.add(Dense(12, activation='softmax'))
model_res.summary()

# Compile the model
model_res.compile(loss=categorical_crossentropy,
              optimizer=Adam(learning_rate=0.0001),
              metrics=['accuracy', 'F1Score', 'Recall', 'Precision'])


n_epochs = 50
batch_size = 32
verbosity = 1
validation_split = 0.30

# Fit data to model
history_res = model_res.fit(X_res_train, y_res_train,
              batch_size=batch_size,
              epochs=n_epochs,
              verbose=verbosity,
              validation_split=validation_split)
````

````python
plot_history(history_res)
# Generate generalization metrics
score = model_res.evaluate(X_res_test, y_res_test, verbose=0)
print(f"Loss: {score[0]}")
for i in range(1, len(model_res.metrics_names)):
    print(f"* {model_res.metrics_names[i]}: {score[i]}")
````

Luego de este entrenamiento con los datos resampleados usando el mismo modelo, aunque una mayor cantidad de epocas, vemos que desaparece el overfitting que se daba en el caso anterior.

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

#Predict
y_prediction = model_res.predict(X_res_test)
val_pred = np.argmax(y_prediction, axis=-1)
val_pred_test = np.argmax(y_res_test, axis=-1)

#Create confusion matrix and normalizes it over predicted (columns)
order_malwares =  le_cat.inverse_transform(np.unique(y_cat))
result = confusion_matrix(val_pred_test, val_pred)
print("Reporte de clasificación:")
print()
print(classification_report(val_pred_test, val_pred,
                            target_names=order_malwares))
print()
````

````python
fig, ax = plt.subplots(figsize=(10, 7))
short_names = ['Adwa', 'Back', 'File', 'PUA', 'Rans', 'Risk', 'Scar', 'Troj', 'TBan', 'TDro', 'TSMS', 'TSpi']
disp = ConfusionMatrixDisplay(confusion_matrix=result, display_labels=short_names)
disp.plot(ax=ax);
````

Tanto la matriz de confusion como el reporte de metricas muestran commo el resampleo de datos ayuda a la igualar las metricas para todas las clases, pero en especial para las que menos datos tenian: FileInfector y Trojan_Banker. Sin embargo el rendimiento de las clases mayoritarias como Riskware y Adware ahora es ligeramente menor al modelo con sobreajuste (Adware ahora da 0.82 y Riskware 0.83 en valores de F1-score).
